```python
class Price:

    def __init__(self,
                 ticker,
                 start=-6106060800, # 1776-7-4 # independence day
                 end=dt.today().timestamp(),
        ):
        '''
        Yahoo Finance
        '''
        self.ticker = ticker.upper()
        self.p1 = int(start)
        self.p2 = int(end)
        
        # price data
        self.raw = self._get_raw()
        self.adj = self._get_adj()
        
        # other non-price data
        self.other = self._get_other()
    
    def __repr__(self):
        
        return f'Yahoo `{self.ticker}` Data'
    
    def _get_raw(self):
        '''
        Get raw data from Yahoo Finance
        '''
        # get from YAHOO finance
        href = f'https://query1.finance.yahoo.com/v7/finance/download/{self.ticker}?period1={self.p1}&period2={self.p2}&interval=1d&events=history'
        df = pd.read_csv(href)
        
        # set index
        df['Date'] = pd.to_datetime(df['Date'])
        df.set_index('Date', inplace=True)
        
        return df
    
    def _get_adj(self):
        '''
        Return data adjusted for stock splits
        '''
        # calculate stock split multiple
        multiple = self.raw['Adj Close'] / self.raw['Close']
        
        # multiply row wise to get adjusted
        df = self.raw.multiply(multiple, axis=0)
        
        # rename organize columns
        self._vol = df['Volume'] # store adjusted volumne
        cols = ['Open', 'High', 'Low', 'Close'] # remaining columns
        df = df[cols]
        
        df['Average'] = df[cols[:-1]].mean(axis=1)
        
        return df
    
    def _get_other(self):
        '''
        Return other non-price data
        '''
        
        res = pd.DataFrame(self._vol)
        
        price = self.adj
        
        res['High-Low'] = price['High'] - price['Low']
        res['Close-Open'] = price['Close'] - price['Open']
    
        return res
```

```python
# return annualization formulas

# method names
ann_keys = ['simp', 'cagr', 'cont']

# corresponding formulas
ann_formulas = [
    
    # simple multiple
    lambda r, p: r * 365 / p,
    
    # annual compounding
    lambda r, p: ((1 + r)**(365 / p)) - 1,
    
    # continous compounding
    lambda r, p: np.exp(r * 365 / p, dtype=np.float64) - 1,
]

# create dictionary to match key and functions
annualizations = {k:v for k,v in zip(ann_keys, ann_formulas)}

# set True as simple annualization (as a sort of default key)
annualizations[True] = ann_formulas[0]

# function to quickly get formulas with either string or index key
get_ann = lambda key: ann_formulas[key] if type(key)==int else annualizations[key[:4].lower()]
```

```python
# calculate returns

def pct_change(self,
               period=1, # return period (days)
               annualize=False,
               **kwargs):
    '''
    Shortcut for calculating percent change / price returns
    '''
    
    # set dataframe
    if isinstance(self, Price):
        df = self.adj.copy()
    else:
        df = self.copy()

    # save original index
    index = df.index

    # resample daily for full days (back filled)
    df = df.resample('D').bfill()
    df = df.reindex(index)

    # replace zeroes to prevent dividing by zero
    cols = df.columns[4:]
    df[cols] = df[cols].replace(0, np.nan)
    df[cols] = df[cols].fillna(method='bfill')
    
        
    # percentage change
    df = df.pct_change(period, **kwargs)

    # return with original index dates
    df = df.reindex(index).dropna().replace(np.inf, 0)

    # annualize
    if annualize:
        df = df.apply(get_ann(annualize), args=(period,))

    return df

# add to object
Price.pct_change = pct_change
```

```python

# stock summary and plots
def summary(self,
            col='Average',
            period=14, # return period
            end=dt.today(), # end date of observations
            n=3*365, # days of observations
            bins=25, # histogram bins
            pct=True, # percentage data (instead of price)
            show=True # plt.show (instead of return)
           ):
    '''
    Summarize stock
    '''

    # data series
    if pct:
        # adjusted return
        df = self.pct_change(period)[col]
    else:
        # adjusted price
        df = self.price[col]

    
    
    # resample daily for full days (back filled)
    index = df.index
    df = df.resample('D').bfill()
    
    # time period
    df = df[df.index <= end][-n:]
    
    # reset to original index dates
    df = df.reindex(index)

    # summary statistics
    stats = pd.DataFrame(
        {'median' : df.median(),
         'var'    : df.var(),
         'skew'   : df.skew(),
         'kurt'   : df.kurt(),
         'start'  : df.index[0],
         'end'    : df.index[-1],
         'days'   : n,
        }, index=[self.ticker])
    
    
    # title
    title = f'{self.ticker}: Adj {col}'
    if pct:
        stats['period'] = period
        title += f' - {period} Day Return'
    else:
        title += 'Price'

    desc = df.describe()
    res = pd.DataFrame(desc.append(stats.T[self.ticker]),
                       columns=[self.ticker])
    
    if show:
        
        # show stats numbers
        display(res.T)
        
        # plot returns
        fig, axes = plt.subplots(2, figsize=(8, 10))

        axes[0].set_title(title,
                      fontsize=18, fontweight='bold')

        kwargs = dict(alpha=0.65)
        axes[0].plot(df, lw=1.5, **kwargs) # plot data
        axes[1].hist(df, bins=bins, **kwargs) # histogram

        # some stats
        mean = desc['mean']
        median = stats['median'][0]
        std = desc['std']

        # mean
        axes[0].axhline(mean, ls='--', c=dcolors[1], label="mean", **kwargs)
        axes[1].axvline(mean, ls='--',  c=dcolors[1], label="mean", **kwargs)

        # median
        axes[0].axhline(median, c=dcolors[2], label="median", **kwargs)
        axes[1].axvline(median, c=dcolors[2], label="median", **kwargs)
        

        # standard deviations
        kwargs = dict(alpha=0.25)
        for i in range(1, 4):
            for s in [std, -std]:
                axes[0].axhline(mean+i*s, ls='--', c='gray', label="std dev", **kwargs)
                axes[1].axvline(mean+i*s, ls='--', c='gray', label="std dev", **kwargs)

        plt.show()
        
    else:
        return res

# add to object
Price.summary = summary
```

```python
class Screener:
    
    def __init__(self,
                 url=None,
                 caps=('Mega', 'Large', 'Mid', 'Small'),
                 limit=None,
                 headless=True
        ):
        
        '''
        Yahoo screener
        '''
        
        self.caps = (caps,) if type(caps)==str else caps
        self.limit = limit
        self.headless = headless
        
        
        
        self.driver, self.url = self.init_driver()
        
        self.raw = dict()
        self.data = None
        
        self._notcomplete = True
        while self._notcomplete:
            try:
                self.get_data()
            except:
                pass
            
        self.driver.close()
            
    
    def __repr__(self):
        return 'Yahoo Stock Screener'
        
    
    def init_driver(self):

        # get url with session key
        print('> Initializing chromedriver')

        # headless Chrome
        chromeOptions = webdriver.chrome.options.Options()
        if self.headless:
            chromeOptions.add_argument("--headless")

        
#         #install driver
#         chromedriver = ChromeDriverManager().install()
        
        # start driver
        driver = webdriver.Chrome(chromedriver, options=chromeOptions)
        

        
        # go to screener page
        try:
            url = 'https://finance.yahoo.com/screener/new'
            driver.get(url)

            # select cap size
            for cap_size in driver.find_elements_by_xpath('//button[contains(text(),"Cap")]'):
                if any(c.lower() in cap_size.text.lower() for c in self.caps):
                    cap_size.click()
                    print('\t> Selected:', cap_size.text)

            # click find stocks
            driver.find_element_by_xpath('//button//*[contains(text(),"Find ")]/..').send_keys('\ue00d')

            # get url with session key
            print('\tWaiting...')
            while url == 'https://finance.yahoo.com/screener/new':
                url = driver.current_url

        # if error show error
        except Exception as e:
            print(e)

        print('\n> url:', url)
        
        return driver, url


    def get_data(self, count=250):

        driver = self.driver

        # check offset, starting at 0 by default
        if not self.raw:
            offset = 0
        else:
            first_key = list(self.raw.keys())[0]
            offset = len(self.raw[first_key])

        url = driver.current_url + f"&offset=0&count={count}"
        driver.get(url)
        doc = html.fromstring(driver.page_source)

        # count per page
        count = len(doc.xpath('//table//td/ancestor::tr')) # reset count for max search results

        # initialize result data
        if not self.raw:

            # table headings
            self.ths = []

            # get column headers
            for th in doc.xpath('//table//th'):
                self.ths.append(th.text_content())
                self.raw[th.text_content()] = list() # set empty list for each header

        first_key = list(self.raw.keys())[0]

        # number of results
        try:
            res_cnt = doc.xpath('//span[contains(text(), "of")][contains(text(), "results")]')[0].text.split()[2]
            res_cnt = int(res_cnt)
        except Exception as e:
            print('\n> Error: check maximum entries\n\n', e)
            return

        # reset limit
        self.limit = res_cnt if self.limit==None else self.limit

        print('Total Results:', res_cnt,'\n')

        # number of pages
        current_cnt = len(self.raw[first_key])
        pages = round((res_cnt-current_cnt)/count)
        t0 = time.time()

        try:
            # iterate pages
            for i in range(pages):

                # check url
                driver.get(self.url + f'''&count={count}&offset={offset}&sort="Market%20Cap"''')
                doc = html.fromstring(driver.page_source)

                if i % 10 == 0 and i>0:

                    # clear progress
                    hlp.clear()

                    d_t = time.time()-t0

                    print(f'> Yahoo Screener | url: {driver.current_url}')
                    print('Elapsed: ', hlp.timer(d_t))
                    r_t = (d_t/(i+1)) * (pages - (i+1))
                    print('Remaining: ', hlp.timer(r_t))
                    print('...')

                current_cnt = len(self.raw[first_key])
                print(f'Page {i+1} of {pages} ({current_cnt} of {res_cnt} Results)')

                # stop if reached limit
                if i>0 and current_cnt == prev_cnt:
                    print('\n> No additional entries')
                    break

                # if meet limit
                if current_cnt >= self.limit:
                    break

                # for table row
                for tr in doc.xpath('//table//tr'):

                    # for table data in row 
                    for j, td in enumerate(tr.xpath('.//td')):

                        # data to header indexed dictionary
                        col_data = self.raw[self.ths[j]]

                        col_data.append(td.text_content())

                        self.raw[self.ths[j]] = col_data

                # update offset
                prev_cnt = current_cnt
                offset += count

        except KeyboardInterrupt:
            print('\n> Interrupted')

        # make sure same size
        vmin = min(len(v) for v in self.raw.values())
        for k, v in self.raw.items():
            self.raw[k] = v[:vmin]

        # to dataframe
        df = pd.DataFrame(self.raw).drop_duplicates()
        df = df.replace('N/A', np.NaN)

        # convert string magnitudes
        df['mktcap'] = df['Market Cap'].apply(hlp.mag_to_int)
        df['Avg Vol (3 month)'] = df['Avg Vol (3 month)'].apply(hlp.mag_to_int)
        df['Volume'] = df['Volume'].apply(hlp.mag_to_int)

        # convert strings to float
        df['Price (Intraday)'] = df['Price (Intraday)'].apply(hlp.str_to_float)
        df['Change'] = df['Change'].apply(hlp.str_to_float)
        df['% Change'] = df['% Change'].apply(hlp.str_to_float) / 100
        df['PE Ratio (TTM)'] = df['PE Ratio (TTM)'].apply(hlp.str_to_float)

        df = df.sort_values('mktcap', ascending=False)

        self.data = df.reset_index(drop=True)

        # calculate implied shares outstanding
        self.data['shares'] = self.data['mktcap'] / self.data['Price (Intraday)']
        
        self._notcomplete = False
```

```python
def data_to_sql(update=True,
                allow_pause=False):

    # load sql
    yahoo_engine = hlp.load_db('yahoo')
    
    # update stock screener
    if update:
        
        # get yahoo screener data
        try:
            scn = Screener()
            scn.get_data()

        except Exception as e:
            print(e)

        # load SQL engine
        yahoo_engine = hlp.load_db('yahoo')

        # save to SQL
        scn.data.to_sql(name='screener',
                        con=yahoo_engine,
                        if_exists='replace',
                        index=False,
                        chunksize=1000)
        tickers = scn.data['Symbol']
        
    # load stock screener
    else:
        yahoo_screener = pd.read_sql('SELECT * FROM screener', con=yahoo_engine)
        tickers = yahoo_screener['Symbol']

        
    # ticker data
    t0 = time.time()

    data_len = len(tickers)

    for i, ticker in enumerate(tickers):

        try:
            # load stock
            price = Price(ticker)

            # save adj price to database
            price.adj.to_sql(
                name=ticker.lower(), 
                con=yahoo_engine,
                if_exists='replace',
                index=True,
                index_label='Date',
                chunksize=1000
            )

            # print progress
            if i%25==0:
                
                hlp.clear_output()
                total = time.time()-t0
                avg_time = total / (i+1)

                print(f'Total time: {hlp.timer(total)} | [{i+1} of {data_len}]')
                print(f'Avg time: {hlp.timer(avg_time)}')
                print(f'ETA: {hlp.timer(avg_time*(data_len-i-1))}')
                print('-'*25, '\n')

            print(f'[{i+1} of {data_len}] {ticker}')

        # pause or kill
        except KeyboardInterrupt:

            if allow_pause:

                print('\n> Paused')
                print('> Interrupt again to continue\n')

                try:
                    while True:
                        pass

                except KeyboardInterrupt:
                    pass

            else:
                print('\n> Break')
                break

        except Exception as e:
            print('\n> Error:\n')
            print(e)
            break
```
